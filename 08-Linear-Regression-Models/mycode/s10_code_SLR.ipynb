```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Using np.polyfit to explore simple linear regression 

```python
df = pd.read_csv("../DATA/Advertising.csv")
```

```python
df.head()
```

```python
# creating new feature 'total_spend' 
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
# looking at the relationship between total advertising spending and sales
sns.scatterplot(data = df, x = 'total_spend', y= 'sales')
```

```python
sns.regplot(data = df, x = 'total_spend', y= 'sales')
```

```python
# declaring our independent and dependent variables: 

X  = df['total_spend']
y = df['sales']
```

```python
#help(np.polyfit)

np.polyfit(X,y, deg = 1)

#B1 = .05, B0 = 4.24 
#y_hat = .05x + 4.24
```

```python
# 500 made up potential spend values
potential_spend = np.linspace(0,500,100)
```

```python
# sales predictions for the spend values above using our linear regression equation parameters B0 and B1 
predicted_sales = 0.04868788*potential_spend + 4.24302822
```

```python
sns.scatterplot(x = 'total_spend', y = 'sales', data = df)
plt.plot(potential_spend, predicted_sales, color = 'red')
```

```python
# how about equation of higher order ? 

np.polyfit(X,y,3)
```

```python
pot_spend = np.linspace(0,500,100)
pred_sales = 3.07615033e-07*pot_spend**3 + -1.89392449e-04*pot_spend**2 +  8.20886302e-02*pot_spend + 2.70495053e+00
```

```python
sns.scatterplot(x = 'total_spend', y = 'sales', data = df)
plt.plot(pot_spend, pred_sales, color = 'red')
```
